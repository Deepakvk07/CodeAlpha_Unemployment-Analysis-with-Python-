```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Unemployment in India.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.strip()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Region'].unique()
```

```python
df['Estimated Unemployment Rate (%)'].mean()
```

```python
df['Estimated Unemployment Rate (%)'].max()
```

```python
df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values().head()
```

```python
df.groupby('Region')[['Estimated Employed','Estimated Unemployment Rate (%)']].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.lineplot(x='Date',y='Estimated Unemployment Rate (%)',data=df)
plt.title('Estimated Unemployment Rate')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Estimated Unemployment Rate (%)',data=df)
plt.title('Estimated Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate')
```

```python
covid_df=df[df['Date'] >= '2020-03-01']
covid_df.head()
```

```python
covid_df.groupby('Region')[['Estimated Employed','Estimated Unemployment Rate (%)']].mean()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Date',y='Estimated Unemployment Rate (%)',data=covid_df)
plt.title('Estimated Unemployment Rate during Covid 19')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
region_data = covid_df.groupby('Region').agg({
    'Estimated Employed': 'mean',
    'Estimated Unemployment Rate (%)': 'mean'
}).reset_index()

region_data['Estimated Employed'] = (region_data['Estimated Employed'] / 1000000)

plt.figure(figsize=(12,6))

plt.plot(region_data['Region'],region_data['Estimated Employed'],marker='o',label='Employed (Millions)')

plt.plot(region_data['Region'],region_data['Estimated Unemployment Rate (%)'],marker='s',label='Unemployment Rate (%)')

plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Value')
plt.title('Employment vs Unemployment After COVID-19 (Region-wise)')
plt.legend()
plt.grid(True)

plt.show()
```
